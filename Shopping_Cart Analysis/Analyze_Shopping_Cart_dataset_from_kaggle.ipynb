# Merge the data

```python
df_merged_1 = pd.merge(order, customer, on='customer_id')
```

```python
df_merged_1.head()
```

```python
df_merged_1.info()
```

```python
df_merged_2 = pd.merge(df_merged_1, sales, on='order_id')
```

```python
df_merged_2.head()
```

```python
df_merged_2.info()
```

```python
df_merged_3 = pd.merge(df_merged_2, product, left_on='product_id', right_on='product_ID')
```

```python
df_merged_3.head()
```

```python
df_merged_3.info()
```

# EDA

```python
# what is the unique values in each categorical column
categorical = df.select_dtypes(["category", "object"]).columns.to_list()
categorical
```

```python
# convert to datetime format
df['order_date'] = pd.to_datetime(df['order_date'])
df['delivery_date'] = pd.to_datetime(df['delivery_date'])
```

```python
# what is the maximum number of days between the order and delivery
df['differance'] = df['delivery_date']- df['order_date']
```

```python
df.head()
```

```python
df['differance'].sort_values(ascending=False).head(20)
```

```python
# extract year , month and day name for order data
df['year_order'] = df['order_date'].dt.year
df['month_order'] = df['order_date'].dt.month
df["day_order"] = df["order_date"].dt.day
df["name_day_order"] = df["order_date"].dt.day_name()
```

```python
df[['year_order', 'month_order', 'day_order', 'name_day_order']].head()
```

```python
# extract year , month and day name for delivery data
df['year_delivery'] = df['delivery_date'].dt.year
df['month_delivery'] = df['delivery_date'].dt.month
df["day_delivery"] = df["delivery_date"].dt.day
df["name_day_delivery"] = df["delivery_date"].dt.day_name()
```

```python
df[['year_delivery', 'month_delivery', 'day_delivery', 'name_day_delivery']].head()
```

```python
# wat is the total money that each gendre spent
gender = df.groupby('gender')[['total_price']].sum().sort_values(by='total_price', ascending = False)
```

```python
px.bar(gender, x=gender.index, y=gender.total_price)
```

```python
# what is the most day that get highest income
day = df.groupby('name_day_order')[['total_price']].sum().sort_values(by='total_price', ascending = False)
```

```python
px.bar(day, x=day.index, y=day.total_price)
```

```python
# what is the highest month that get the highest sales?
month = df.groupby('month_order')[['total_price']].sum().sort_values(by='total_price',ascending = False)
```

```python
px.bar(month, x=month.index, y=month.total_price)
```

```python
# what is the highest state achieve the highest salees
state = df.groupby(['state'])[['total_price']].sum().sort_values(by='total_price', ascending = False)
```

```python
px.bar(state, x=state.index, y=state.total_price)
```

```python
# what is the highest state per gender achieve the highest sales
s_g = df.groupby(['state','gender'])[['total_price']].sum().sort_values(by='total_price',ascending = False).reset_index()
```

```python
px.bar(s_g, x=s_g['state'], y=s_g['total_price'], color=s_g['gender'])
```

```python
# what is the average age per state
avr_age = df.groupby(['state'])[['age']].mean().sort_values(by='age', ascending = False)
```

```python
px.bar(avr_age, x=avr_age.index, y= avr_age.age)
```

```python
# what is the average age per state per each gender
age = df.groupby(['state','gender'])[['age']].mean().sort_values(by='age',ascending = False).head(30).reset_index()
```

```python
px.bar(age, x=age['state'], y=age['age'], color=age['gender'])
```
